```python
#One-hot-encoding
# Display the original dataset
df_1 = pd.read_csv(path)
print("Original Dataset:")
print(df_1)

# One-hot encode categorical columns
df_1 = pd.get_dummies(df_1, columns=['cap-shape',
                                 'cap-surface',
                                 'cap-color',
                                 'bruises%3F',
                                 'odor',
                                 'gill-attachment',
                                 'gill-spacing',
                                 'gill-size',
                                 'gill-color',
                                 'stalk-shape',
                                 'stalk-root',
                                 'stalk-surface-above-ring',
                                 'stalk-surface-below-ring',
                                 'stalk-color-above-ring',
                                 'stalk-color-below-ring',
                                 'veil-type',
                                 'veil-color',
                                 'ring-number',
                                 'ring-type',
                                 'spore-print-color',
                                 'population',
                                 'habitat',
                                 'class'])

# Display the dataset after conversion
print("\nDataset after Conversion:")
print(df_1)
```

```python
#Label encoding(Dummy numbers)
# Display the original dataset
df_2 = pd.read_csv(path)

print("Original Dataset:")
print(df_2)

# Iterate through each column in the DataFrame
for column in df_2.columns:
    # Check if the data type of the column is 'object' (string)
    if df_2[column].dtype == 'object':
        # Convert string values to integers
        df_2[column] = pd.Categorical(df_2[column])
        df_2[column] = df_2[column].cat.codes

# Display the dataset after conversion
print("\nDataset after Conversion:")
print(df_2)
```

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

X = df_2.drop(["class"], axis=1).values
y = df_2["class"].values



X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30, shuffle=True)


clf = RandomForestClassifier(n_estimators = 10)

clf.fit(X_train, y_train)

#pd.set_option('display.max_columns', None)
#print(df_2[1:2])

# Predicts the first two mushrooms
# 1st one is poisonous, 2nd one is edible
#clf.predict([[5, 2, 4, 1, 6, 1, 0, 1, 4, 0, 2, 2, 2, 7, 7, 0, 2, 1, 4, 2, 3, 5], [5, 2, 9, 1, 0, 1, 0, 0, 4, 0, 1, 2, 2, 7, 7, 0, 2, 1, 4, 3, 2, 1]])
```

```python
# Feature importance
feature_imp = pd.Series(clf.feature_importances_, index = df_2.drop(["class"], axis=1).columns).sort_values(ascending = False)
print(feature_imp)
```

```python
from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score

kfold = KFold(n_splits=10)

# Perform k-fold cross validation
for train_index, test_index in kfold.split(X_train):
  X_train_fold, X_test_fold = X_train[train_index], X_train[test_index]
  y_train_fold, y_test_fold = y_train[train_index], y_train[test_index]

  clf.fit(X_train_fold, y_train_fold)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import r2_score
from scipy.stats import linregress

print("Training Data Metrics")
print("---------------------------------------")
#get statistics from training set predictions
y_pred_train = clf.predict(X_train)

#model accuracy
print("Model accuracy : " + str(accuracy_score(y_train, y_pred_train)))
print("\n")

#confusion matrix
confusion_train = confusion_matrix(y_train, y_pred_train)
print("Confusion matrix: ")
print(confusion_train)
print("\n")

#precision and recall
tn, fp, fn, tp = confusion_train.ravel()
print("Model Precision: " + str(tp/(tp+fp)))
print("Model Recall: " + str(tp/(tp+fn)))
print("\n\n")

print("Test Data Metrics")
print("-----------------------------------------------")
#get statistics from test set predictions
y_pred = clf.predict(X_test)

#model accuracy for test data
print("Model accuracy: " + str(accuracy_score(y_test, y_pred)))
print("\n")

#confusion matrix for test data
confusion = confusion_matrix(y_test, y_pred)
print("Confusion matrix: ")
print(confusion)
print("\n")

#precision and recall for test data
tn, fp, fn, tp = confusion.ravel()
print("Model Precision: " + str(tp/(tp+fp)))
print("Model Recall: " + str(tp/(tp+fn)))


slope, intercept, r_value, p_value, std_err = linregress(df_2['odor'].values, df_2['class'].values)
print(r_value)
print(p_value)
```
