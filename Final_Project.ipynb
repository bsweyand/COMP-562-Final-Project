```python
#One-hot-encoding
# Display the original dataset
df_1 = pd.read_csv(path)
print("Original Dataset:")
print(df_1)

# One-hot encode categorical columns
df_1 = pd.get_dummies(df_1, columns=['cap-shape',
                                 'cap-surface',
                                 'cap-color',
                                 'bruises%3F',
                                 'odor',
                                 'gill-attachment',
                                 'gill-spacing',
                                 'gill-size',
                                 'gill-color',
                                 'stalk-shape',
                                 'stalk-root',
                                 'stalk-surface-above-ring',
                                 'stalk-surface-below-ring',
                                 'stalk-color-above-ring',
                                 'stalk-color-below-ring',
                                 'veil-type',
                                 'veil-color',
                                 'ring-number',
                                 'ring-type',
                                 'spore-print-color',
                                 'population',
                                 'habitat',
                                 'class'])

# Display the dataset after conversion
print("\nDataset after Conversion:")
print(df_1)
```

```python
#Label encoding(Dummy numbers)
# Display the original dataset
df_2 = pd.read_csv(path)

print("Original Dataset:")
print(df_2)

# Iterate through each column in the DataFrame
for column in df_2.columns:
    # Check if the data type of the column is 'object' (string)
    if df_2[column].dtype == 'object':
        # Convert string values to integers
        df_2[column] = pd.Categorical(df_2[column])
        df_2[column] = df_2[column].cat.codes

# Display the dataset after conversion
print("\nDataset after Conversion:")
print(df_2)
```

```python
# visualization
import seaborn as sns
import matplotlib.pyplot as plt

img_path = 'imgs'
if not os.path.exists(img_path):
    os.makedirs(img_path)

# show every graph in one page
i=1
plt.figure(figsize=(30, 48))
for column in df.columns[:-1]:
    plt.subplot(8,3,i)
    # plt.figure(figsize=(10, 6))
    sns.countplot(y=column, hue="class", data=df, palette=['crimson', 'mediumspringgreen'])
    plt.xlabel("count",fontsize=20)
    plt.ylabel(column,fontsize=20)
    plt.legend(fontsize=20)
    plt.tick_params(axis='both', labelsize=16)
    plt.title(f'Distribution of {column} by edibility',fontsize=20)
    i+=1
name=img_path+"/all"
plt.subplots_adjust(left=0.05, right=0.95, top=0.95, bottom=0.05, hspace=0.6, wspace=0.2)
plt.savefig(name)
plt.show()
plt.close()

# save each graph in different file
# for column in df.columns[:-1]:
#     plt.figure(figsize=(10, 6))
#     sns.countplot(y=column, hue="class", data=df, palette=['crimson', 'mediumspringgreen'])
#     plt.xlabel("count",fontsize=20)
#     plt.ylabel(column,fontsize=20)
#     plt.title(f'Distribution of {column} by edibility',fontsize=20)
#     plt.legend(fontsize=20)
#     plt.tick_params(axis='both', labelsize=16)
#     name=img_path+"/"+column
#     plt.savefig(name)
#     plt.close()
```
